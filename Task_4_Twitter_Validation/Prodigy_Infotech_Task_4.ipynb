```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("twitter_validation.csv")
```

```python
df = data.groupby(['Company', 'Response'])['S_No'].count().reset_index()
```

```python
data.columns
```

```python
df
```

```python
df.sort_values(by='S_No',ascending=False,inplace=True)
df
```

```python
df.Response.unique()
```

```python
s = ['Neutral','Irrelevant', 'Positive','Negative']
```

```python
plt.figure(figsize=(20,40))

# Increase the aspect ratio to make the bars wider
sns.set_theme(style="whitegrid")
ax = sns.barplot(data=df, x='S_No', y="Company", hue="Response",capsize= 4, linewidth=5, errcolor=".5",hue_order=s)


plt.yticks(fontsize=10)
plt.tick_params(axis='x', labelsize=20, rotation=0)
plt.tick_params(axis='y', labelsize=20, rotation=0)
plt.xlabel('Response', fontsize=25)
plt.ylabel('Company', fontsize=25)
plt.legend(loc='center right', fontsize=20)

plt.show()
```
